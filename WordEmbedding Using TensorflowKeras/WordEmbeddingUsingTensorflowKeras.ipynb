```
from tensorflow.keras.preprocessing.text import one_hot
```

```
#Sentences
sentences = ["I am Mohit",
             "I am 22 years old",
             "I graduated from jims",
             "I love to learn new things everyday",
             "My hobbie is to doing gymming"]
```

```
sentences
```

```
#Vocabulary Size
voc_size = 10000
```

```
#Fetching the index of each word from vocabulary using one-hot
#One-hot encodes a text into a list of word indexes of size n

wordIndex = [one_hot(word,voc_size) for word in sentences]
wordIndex
```

```
#form embedding matrix by passing wordIndex in Embedding keras layers 

from tensorflow.keras.preprocessing.sequence import pad_sequences #Use the make all the sentences have same number of words that help us to create very god embedding matrix
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
```

```
#It converts every sentence to the 10 word sentences
sent_length = 10
embeddedWords = pad_sequences(wordIndex,maxlen=sent_length,padding='pre')
embeddedWords
```

```
#Now passing the embedded words into the embedding layers
dimensions = 15                           #Dimensions are the no. of features for every word
model = Sequential()
model.add(Embedding(voc_size,dimensions,input_length=sent_length))
model.compile(optimizer='adam',metrics="mse")             #Compile with adam & Mean Squared Error
```
